## Background

I was puzzled when I first heard the instructor mention moments in the lesson about statistics. My background is in mechanical engineering, so when I hear moment, I think of torque or moment of inertia. 

Upon researching what a moment is in statistics, I found this helpful article, which makes a brief mention that the two types of moments are actually related. To put it abstractly, a moment of a function (statistical, physical, or otherwise) is a way of describing the shape or distribution of data relative to a point.

For example, torque (the product of a force and it's distance from a point) is the first moment of force and it quantifies the relationship of that force about the chosen point.

In statistics, each moment (categorized in a numerical hierarchy called orders) describes the statistical distribution of data. A distribution is just a function that explains the possible values for a variable and how frequently that value occurs. There are many different common types of distributions, such as Normal, Binomial, and Uniform distributions.

Each of the four moments is derived from a simple equation:

(x1^s + x2^s + … + xn^s) / n

where x1, x2, etc. are data points generated by the distribution function for n discrete data points. Each data point is raised to an exponent s and then the sum of these points are divided by the count of all the data points. The exponent s is the order of the moment.

## 1. Mean

The mean (μ)- also called the average - is the most commonly understood moment. It is the first moment of a distribution, meaning that the exponent s is equal to 1.

If we had a set of three data points (n = 3), then the first moment equation for the mean (μ) would look like this:

μ = (x1¹ + x2¹+ x3¹) / 3 = (x1 + x2 + x3) / 3

In python you can use the Numpy library to calculate the mean of a set of data like this:

```python
import numpy as np
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis, shapiro

dataset = [1, 2, 3]
print(np.mean(dataset))
```

A financial application of this would be to find the average daily return of a stock.

```python
symbol = "AAPL"
obj = yf.Ticker(f"{symbol}")
price_history = obj.history(period = "1y", auto_adjust = True, interval = "1d").dropna()["Close"]
daily_returns = price_history.pct_change().dropna()
average_daily_return = np.mean(daily_returns)
print(f"The average daily return over the past year is {round(average_daily_return*100,3)}%")

sns.set_style("darkgrid")
ax = sns.histplot(daily_returns*100, element="poly")
plt.axvline(average_daily_return, 0,1)
ax.set(xlabel='Daily Return (%)', ylabel='Days Per Return', title='Distribution of Daily Returns (1Y)')
```

This plot shows the distribution of daily returns over the past ten years, where the horizontal axis shows the daily return as a percentage, and the vertical axis shows how many days returned that percentage.

However, the average daily return does not take one important detail into consideration - compound interest. If you buy and hold AAPL, your return on the investment is going to depend on how long you hold it for, since each day is going to compound - otherwise you would only get a measly 0.143% return.

There are three main ways to include compounding in calculating your return: annualized average return, annualized total return, and compound annual growth rate.

The annualized average return uses the average daily return to approximate what the annual return would be with compounding. The DataCamp tutorial suggests to annualize the return using the following equation given an average of 252 trading days in a year.

```python
annualized_average_return = ((1+np.mean(daily_returns))**252)-1
print(f"The annualized average return is {round(annualized_average_return*100,3)}%")
```

2. The Corporate Finance Institute provides the following equation for calculating the annualized total return when you know the cumulative return R for the number of days held n:

```python
R = 0.001
n = 366
annualized_total_return = ((1+R)**(365/n) - 1)*100
print(f"The annualized total return is {round(annualized_total_return*100,3)}%")
```

3. Lastly, there is the compound annual growth rate (CAGR). Instead of using the average daily return, the CAGR uses the ratio of the price at the end of a period to the price at the start. This is raised to the power of the ratio of the unit time period to the holding period length. In the example below, the unit time period is one market year, expressed as 252 days. The denominator - the holding period length - is the length of the list of price history data, which also happens to be 252 days since we requested 1 market year of data.

```python
CAGR = ( (price_history[-1] / price_history[0]) ** (252/len(price_history)) -1) *100
print(f"The CAGR is {round(CAGR,3)}%")
```

The CAGR is closer to the first calculation of average annualized return, but it is still off slightly because of the difference in return calculation (last:first vs. average daily return).

## 2. Variance

Before I get into variance, I want to explain standard deviation σ (this is because variance is the standard deviation squared). In finance, the standard deviation of an asset's return is often referred to as volatility.

The standard deviation can be a bit of a convoluted concept to understand, so I think it's easiest to break it down into layers or steps:

1. We have a data set, such as a list of historical price records.

2. We can find the average value of the data set by calculating the mean μ.

3. Let's say we were to find how far every data record deviated from the average. For every data record, we would subtract the data point from the average value to find this distance.

4. While we're at it, let's also square each of these deviations (don't ask me why).

5. This part will look familiar. We are going to find the average of the squared deviations by adding all of them up and then dividing by the quantity of deviations. This is the variance σ², the second moment of a statistical distribution.

6. Lastly, we are going to take the square root of this average squared deviation. This is the standard deviation σ.

You can think of standard deviation as a way of organizing groups of data around the mean. In a normal distribution (more on normality later), about 68% of all the data in the dataset will fall within +/- 1σ. So, if the average daily return of AAPL is 0.143% and the standard deviation is:

```python
std_dev_daily = np.std(daily_returns)
print(f"The daily volatility is +/-{round(std_dev_daily*100,3)}%.")
```

Then there is a 68% chance that the daily average return will be between -1.632% and 1.918% (the useful thing about standard deviation is that it is in the same units as the data). The standard deviation also tells us the range since - in a normal distribution - about 99.8% of all data will fall within +/- 3σ. This means that a stock with a relatively high standard deviaion of returns is more risky and potentially more rewarding than one with a low standard deviation of returns.

The standard deviation of the returns (volatility) can be scaled with the square root of time to find the volatility across larger time periods, such as monthly (time = 21 trading days) or annualy (time = 252 trading days), like this:

```python
std_dev_monthly = std_dev_daily * (21**0.5)
print(f"The monthly volatility is +/-{round(std_dev_monthly*100,3)}%.")

std_dev_annual = std_dev_daily * (252**0.5)
print(f"The annual volatility is +/-{round(std_dev_annual*100,3)}%.")
```

Now, whereas the standard deviation shows how data can be categorized into groups distributed around the mean and how far the data can deviate, the variance (σ²) gives an indication of the average deviation. The variance does not have the same units as the daily returns or standard deviation (percent in our case), and can be annualized using the annualized standard deviation of returns.

```python
daily_variance = (std_dev_daily)**2
print(f"The daily variance is {daily_variance}.")

annual_variance = (std_dev_annual)**2
print(f"The annualized variance is {annual_variance}.")
```

You may be wondering why you should care about a metric that isn't even measured in units you can understand. 

One of the important aspects of analyzing risk with variance is that the variance is affected more by outliers than by data points close to the mean. This is important for risk analysis because you need to watch out for outliers that can tank your portfolio. 

By using the variance to compare risks of investments, it will be easier to spot an asset that is particularly more volatile than the others. Variance is also important in the analysis of a portfolio's covariance since a portfolio with low covaried assets can expect greater returns for less risk.

Note: before moving on to the third moment, I want to mention two other aspects of statistical analysis. 

One is that there are actually different equations for sample analysis and population analysis, where a sample is a small part of a larger population of data. 

The other detail to consider is a metric called the coefficient of variation, which is a way of directly comparing the standard deviation of two data sets that have different units, such as comparing variations in different currencies.


## 3 - Skewness

The third moment of a distribution, called skewness, is a measure of the tilt of the distribution towards either tail end of the curve. 

In a normal distribution, skewness is near zero, meaning that the curve is mostly symmetric about the mean. However, financial returns are often not normally distributed, tending towards positive skewness. 

Although there are more precise ways of evaluating normality, observing a significant skewness is often a reasonable indication of non-normality.

Positive skewness means that the distribution leans towards the left tail and negative skewness means that it leans towards the right tail. According to Dakota Wixom (the instructor of the DataCamp course I am taking), positive skewness is desired in finance because it suggests a "higher probability of significantly good returns on the right hand side of the distribution, and a compressed, predictable left-hand distribution of negative returns."

Consider the distribution of daily returns below. The average return is greater than the median return, which is a characteristic of positive skewness.

```python
sns.set_style("darkgrid")
ax = sns.histplot(daily_returns*100, element="poly")
plt.axvline(average_daily_return, 0,1)
ax.set(xlabel='Daily Return (%)', ylabel='Days Per Return', title='Distribution of Daily Returns (1Y)')

median = np.median(daily_returns)
print(f"The median daily return is {round(median*100,3)}%.")
print(f"The average daily return is {round(average_daily_return*100,3)}%.")
print(f"The difference between the average and the median daily return is {round((average_daily_return - median)*100, 3)}%.")

skewness = skew(daily_returns.dropna())
print(f"\nThe skewness of the daily returns is {skewness}.")
```

## 4. Kurtosis

The fourth moment of a distribution, kurtosis, quantifies the thickness of the tail ends of the distribution. This useful metric can be used as an indication of the probabilty of an outlier event occurring, such as an extreme positive or negative return. A normal distribution will have a kurtosis nearly equalt to three. Typical financial returns are leptokurtic, which just means that the distribution has excess kurtosis (greater than three). A kurtosis value of three is used as a baseline value, where any distribution with a deviation from this baseline has a higher (leptokurtic) or lower (platykurtic) probability of an outlier occuring than normal.

We can use kurtosis() from scipy.stats to calculate excess kurtosis. If we want to calculate the sample kurtosis, we need to add three to the excess kurtosis.

```python
excess_kurtosis = kurtosis(daily_returns.dropna())
print(f"The excess kurtosis of the daily returns distribution is {round(excess_kurtosis,3)}")

sample_kurtosis = excess_kurtosis + 3
print(f"The sample kurtosis of the daily returns distribution is {round(sample_kurtosis,3)}")
```

Although it might not be obvious, the distribution has a slight positive skewness, which we can calculate using skew() from the scipy.stats package. Since the distribution has positive skewness, a majority of the returns data fall to the right of the mean, meaning that a majority of the returns are greater than average.

## 5. Normality

When looking at a distribution that has a non-zero skewness and leptokurtosis, one could assume that the data is non-normal; however, it would be better to prove this with a test for normality, such as the Shapiro-Wilk test.

The Shapiro-Wilk test has a null hypothesis of normality. A null hypothesis is essentially a hypothesis that two sets of data are not different in regards to a particular characteristic, and is assumed to be true until proven otherwise. In the case of the Shapiro-Wilk test, you are testing whether or not your sample data set differs from a normal distribution.

We can use shapiro() from scipy.stats to test the returns distribution for normality. It will return a t-stat and a p-value. We will focus on the p-value, which represents the probability that the two data sets differ because of chance. 

For the Shapiro-Wilk test, we assume that if the returned p-value is greater than 0.5, then the sample data is likely normally distributed. If the p-value is less than 0.5, then the null hypothesis of normality is rejected, and the alternate hypothesis of non-normality is accepted.

```python
shapiro_results = shapiro(daily_returns)
p_value = shapiro_results[-1]

if p_value >= 0.5:
    print(f"The p-value is {p_value}, indicating normality. Null hypothesis accepted.")
else:
    print(f"The p-value is {p_value}, indicating non-normality. Null hypothesis rejected.")
```
